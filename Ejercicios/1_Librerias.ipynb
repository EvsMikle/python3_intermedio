```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/Salary_Data.csv')
df.head()
```

Comparacion de salarios contra los años de experiencia de la carrera de Software Engineer

```python
x = df.loc[df['Job Title'] == 'Software Engineer', 'Years of Experience']
y = df.loc[df['Job Title'] == 'Software Engineer', 'Salary']

x1 = df.loc[df['Job Title'] == 'Data Analyst', 'Years of Experience']
y1 = df.loc[df['Job Title'] == 'Data Analyst', 'Salary']
```

```python
def demo(sty):
    plt.style.use(sty)
    
    fig, ax = plt.subplots(figsize=(6,6))
    ax.set_title('Años de experiencia vs Salario (Software Engineer)', color='C0')
    
    ax.scatter(x, y, label='Software Engineer')
    ax.scatter(x1, y1, label='Data Analyst')
    plt.legend()
    plt.show()
```

```python
demo('ggplot')
```
